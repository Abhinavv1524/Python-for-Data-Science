### MatplotLib Library


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c="r")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
y=x*x
```

```python
plt.plot(x,y,'go--')
```
